# Detecting Faces with and without masks


```python
from keras.models import load_model
import cv2
import numpy as np
import pandas as pd
import os
```

```python

model = load_model('model-017.model')
face_clsfr=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

source=cv2.VideoCapture(0)

labels_dict={0:'MASK',1:'NO MASK'}
color_dict={0:(255,0,0),1:(0,0,255)}
```

```python
source.set(3, 480) #set width of the frame
source.set(4, 640) #set height of the 
MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746)
age_list = ['(0, 2)', '(4, 6)', '(8, 12)', '(15, 17)','(18,20)', '(21, 23)','(24,25)','(26, 30)','(31, 37)', '(38, 43)', '(48, 53)', '(60, 100)']
gender_list = ['Female','Male', ]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
age_net = cv2.dnn.readNetFromCaffe('age_deploy.prototxt', 'age_net.caffemodel')
gender_net = cv2.dnn.readNetFromCaffe('deploy_gender.prototxt', 'gender_net.caffemodel')
```

```python
while(True):


    ret,img=source.read()
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_clsfr.detectMultiScale(gray,1.3,5) 
    
    
    
    for x,y,w,h in faces:
        
        face_img=gray[y:y+w,x:x+w]
        
        resized=cv2.resize(face_img,(100,100))
        
        normalized=resized/255.0
        
        reshaped=np.reshape(normalized,(1,100,100,1))
        result=model.predict(reshaped)
        
        label=np.argmax(result,axis=1)[0]
        face_img1 = img[y:y+h, h:h+w]
        blob = cv2.dnn.blobFromImage(face_img1, 1, (227, 227), MODEL_MEAN_VALUES, swapRB=False)
        #umang 01/apr
    

        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(0,0,0),2)
        
        if label==1:
            
            #Predict Gender
            gender_net.setInput(blob)
            gender_preds = gender_net.forward()
            gender = gender_list[gender_preds[0].argmax()]
            print("Gender : " + gender)
        
            #Predict Age
            age_net.setInput(blob)
            age_preds = age_net.forward()
            age = age_list[age_preds[0].argmax()]
            print("Age Range: " + age)
            
            overlay_text = "%s %s" % (gender, age)
            cv2.putText(img, overlay_text,(x, w), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
        

        
        
       
       
    
            
    
    #for i in range(100):
        #i =0
        #res = pd.DataFrame(result)
        #res.index = id[i] # its important for comparison
        #res.columns = ["mask",'no mask']
        #res.to_csv("prediction_results.csv")
        #
        
    cv2.imshow('LIVE',img)
    key=cv2.waitKey(1)
    
    if(key==27):
        break
        
# for f in faces:
        
#         if label==1:
#             x=x+1000
#             y=y+1000
#             x, y, w, h = [ v for v in f ]
#             cv2.rectangle(img, (x,y), (x+w,y+h), (255,255,255))
#             #Save just the rectangle faces in SubRecFaces
#             sub_face = img[y:y+h, x:x+w]
#             FaceFileName = "crop/face_"+str(y) + ".png"
#             cv2.imwrite(FaceFileName, sub_face)
#             os.system(" echo 1 >> /root/fortrial/new1.txt")
#             os.system(" git commit -m 'nn' new1.txt ")
#             os.system(" git push ")
    
cv2.destroyAllWindows()
source.release()

```
